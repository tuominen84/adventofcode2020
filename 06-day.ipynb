# Part 1

```python
def load_data(fname):
    with open(fname, 'r') as f:
        data = f.read()
    data = [d.replace('\n', '') for d in data.split('\n\n')]
    
    return data
load_data('inputs/06-test')
```

```python
answers = 0
for g in load_data('inputs/06-input'):
    answers += len(set(g))

print(answers)
```

# part 2

```python
def load_data(fname):
    with open(fname, 'r') as f:
        data = f.read()
    data = [d.split() for d in data.split('\n\n')]
    
    return data
load_data('inputs/06-test')
```

```python
def common_answers(fname):
    common = 0
    for g in load_data(fname):
        print(g)
        #print(set().intersection(*g))
        if len(g) == 1:
            common += len(g[0])
        else:
            common += len(set(g[0]).intersection(*g[1:]))
    return common

common_answers('inputs/06-input')
        
```
