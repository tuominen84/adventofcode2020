```python
def load(fn):
    with open(fn, 'r') as f:
        seats = [l.strip() for l in f.readlines()]
        
    return len(seats), len(seats[0]), list(''.join(seats))
```

```python
rows, cols, seats = load('inputs/11-input')
```

```python
seat_ids = [(r, c) for r in range(rows) for c in range(cols)]
```

```python
ADJ_SEATS = [(x,y) for x in [-1, 0, 1] for y in [-1,0,1]]
ADJ_SEATS.remove((0,0))
ADJ_SEATS
```

```python
def is_seat(s, xseats=seats):
    return xseats[s[0]*cols + s[1]] == '#' or xseats[s[0]*cols + s[1]] == 'L'
def valid_seat(s):
    return s[0] >= 0 and s[0] < rows \
        and s[1] >= 0 and s[1] < cols

def adj_seats(s, all_seats):
    seats_seen = []
    for adj in ADJ_SEATS:
        for t in range(1, max(rows,cols)):
            test_seat_id = (s[0] + adj[0]*t, s[1] + adj[1]*t)
            if not valid_seat(test_seat_id):
                break
            if is_seat(test_seat_id, all_seats):
                #print(test_seat_id, True)
                seats_seen.append(test_seat_id)
                break
    return seats_seen

def compute_adj_seats(seat_ids, all_seats):
    seats_seen = {}
    for s in seat_ids:
        seats_seen[s] = adj_seats(s, all_seats)
    return seats_seen
#adj_seats((6,9), seats)
#is_seat((0,2), seats)
compute_adj_seats(seat_ids, seats)
```

```python

def occupied(s):
    return seats[s[0]*cols + s[1]] == '#'
def empty(s):
    return seats[s[0]*cols + s[1]] == 'L'
def set_seat(s, state):
    new_seats[s[0]*cols + s[1]] = state
```

```python
def print_seats(s):
    for i in range(rows):
        print(''.join(s[i*cols:i*cols + cols]))
```

```python
new_seats = seats.copy()
print(seats != new_seats)
seats_seen = compute_adj_seats(seat_ids, seats)
i = 0
while True:   
    for s in seat_ids:
        adj = seats_seen[s]
        num_occupied = sum([1 for i in adj if occupied(i) ])
        if empty(s) and num_occupied == 0:
            set_seat(s, '#')
        elif occupied(s) and num_occupied >= 5:
            set_seat(s, 'L')
    i += 1
    if new_seats == seats:
        print('tll')
        break
    #print(i)
    #print_seats(new_seats)
    seats = new_seats.copy()
#seats = new_seats.copy()
print(seats != new_seats)
print('occupied', sum([1 for x in seats if x == '#']))
```
