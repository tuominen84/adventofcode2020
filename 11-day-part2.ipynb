```python
def load(fn):
    with open(fn, 'r') as f:
        seats = [l.strip() for l in f.readlines()]
        
    return len(seats), len(seats[0]), list(''.join(seats))
```

```python
rows, cols, seats = load('inputs/11-input')
```

```python
seat_ids = [(r, c) for r in range(rows) for c in range(cols)]
```

```python
ADJ_SEATS = [(x,y) for x in [-1, 0, 1] for y in [-1,0,1]]
ADJ_SEATS.remove((0,0))
ADJ_SEATS
```

```python
def adj_seats(s, all_seats):
    ss = [(s[0] + a[0], s[1] + a[1]) for a in ADJ_SEATS]
    return [i for i in ss if i[0] >= 0 and i[1] >= 0 and i[0] < rows and i[1] < cols]
adj_seats((8,8))
```

```python
def occupied(s):
    return seats[s[0]*cols + s[1]] == '#'
def empty(s):
    return seats[s[0]*cols + s[1]] == 'L'
def set_seat(s, state):
    new_seats[s[0]*cols + s[1]] = state
```

```python
def print_seats(s):
    for i in range(rows):
        print(''.join(s[i*cols:i*cols + cols]))
```
