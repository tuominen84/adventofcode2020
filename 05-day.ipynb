```python
def traverse_bsp(up_down, bsp, rows):
    for d in bsp:
        cut_point = int((sum(rows))/2)
        if d == up_down[0]:
            rows = (rows[0], cut_point)
            print(d, rows)
        else: # d == 'B'
            rows = (cut_point+1, rows[1])
            print(d, rows)
    return rows[0]
traverse_bsp(('F', 'B'), 'FBFBBFB', (0,127))
```

```python
traverse_bsp(('L', 'R'), 'RLR', (0,7))
```

```python
def find_seat(seat_code):
    row = traverse_bsp(('F', 'B'), seat_code[:7], (0, 127))
    col = traverse_bsp(('L', 'R'), seat_code[-3:], (0,7))
    seat_id = row * 8 + col
    #print(row, col, seat_id)
    return row, col, seat_id, seat_code
```

```python
find_seat('FBFBBFFRLR')
```

```python
def check_boardin_passes(passes):
    seats = [ find_seat(p) for p in passes]
    return (max(seats, key=lambda x: x[2]))

check_boardin_passes(load_data('inputs/05-test-input'))
```
